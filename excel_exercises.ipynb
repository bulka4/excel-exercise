```python
# importing neccessary libraries

import pandas as pd
import numpy as np
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
from tqdm import tqdm

import seaborn as sns

import shutil
```

## exercise 1

```python
# load data from the sheet

sheet_data = pd.read_excel('Excel_Zadania_Final.xlsx', 'Excercise1_Zadanie 1')
sheet_data.columns = [i for i in range(len(sheet_data.columns))]
```

```python
sheet_data.head(10)
```

```python
# function to get supplier number from code

def get_supplier_number(code):
    supplier_number = ''

    code_split = code.split('-')
    supplier_number = code_split[0]

    i = 0
    while supplier_number[0] == '0':
        supplier_number = supplier_number[1 : ]
                
    return supplier_number
```

```python
# preparing table with supplier's number column

table = sheet_data.iloc[5 : , 1 : 3]
table.iloc[ : , 1] = table.iloc[ : , 0].apply(lambda x: get_supplier_number(x))
```

```python
table.head(10)
```

```python
# inserting values into excel file

# if there is no file Excel_Zadania_Final_solutions.xlsx then create it (it is a copy of Excel_Zadania_Final.xlsx)
try:
    wb = load_workbook('Excel_Zadania_Final_solutions.xlsx')
except:
    shutil.copyfile('Excel_Zadania_Final.xlsx', 'Excel_Zadania_Final_solutions.xlsx')
    wb = load_workbook('Excel_Zadania_Final_solutions.xlsx')
    
ws = wb.worksheets[2]

for i, value in enumerate(table.iloc[ : , 1]):
    ws.cell(row = 7 + i, column = 3).value = value
    
wb.save("Excel_Zadania_Final_solutions.xlsx")
```

```python
# load data from sheet

sheet_data = pd.read_excel('Excel_Zadania_Final.xlsx', 'Excercise2_Zadanie 2')
sheet_data.columns = [i for i in range(len(sheet_data.columns))]
```

```python
sheet_data.head(10)
```

```python
# preparing table with suppliers number, system id, sequence number, indicator

table = sheet_data.iloc[5 : , 1 : 6]
table.iloc[ : , 1] = table.iloc[ : , 0].apply(lambda x: get_supplier_number(x))
table.iloc[ : , 2] = table.iloc[ : , 0].apply(lambda x: x.split('-')[1])
table.iloc[ : , 3] = table.iloc[ : , 0].apply(lambda x: x.split('-')[2])
table.iloc[ : , 4] = table.iloc[ : , 0].apply(lambda x: x.split('-')[3])
```

```python
table.head()
```

```python
# inserting values into excel file

wb = load_workbook('Excel_Zadania_Final_solutions.xlsx')
ws = wb.worksheets[3]

for i, value in enumerate(table.iloc[ : , 1 : ].values):
    ws.cell(row = 7 + i, column = 3).value = value[0]
    ws.cell(row = 7 + i, column = 4).value = value[1]
    ws.cell(row = 7 + i, column = 5).value = value[2]
    ws.cell(row = 7 + i, column = 6).value = value[3]
    
wb.save("Excel_Zadania_Final_solutions.xlsx")
```

## exercise 3

```python
# load data from sheet

sheet_data = pd.read_excel('Excel_Zadania_Final.xlsx', 'Excercise3_Zadanie 3')
sheet_data.columns = [i for i in range(len(sheet_data.columns))]
```

```python
sheet_data
```

```python
# column for task 2, names without duplicates

names = sheet_data.iloc[8 : , 3]
names.drop_duplicates(inplace = True)
names.reset_index(drop = True, inplace = True)
```

```python
names.head(10)
```

```python
# preparing table for task 3

table1 = sheet_data.iloc[8 : , 1 : 4].copy().reset_index(drop = True)
table2 = pd.DataFrame(sheet_data.iloc[8 : , 5].copy().reset_index(drop = True))

pivot_table = pd.pivot_table(data = table1, index = 1, columns = 2, aggfunc = 'count')[3]
pivot_table = pivot_table.fillna(0).astype(int).reset_index()

table2 = table2.astype(int).set_index(table2.columns[0]).join(pivot_table.set_index(pivot_table.columns[0]))
table2 = table2.reset_index().sort_values('index', ascending = False)

# sorting values in table 2 for task 4
table2 = table2.sort_values(by = table2.columns[0], ascending = False).reset_index(drop = True)
table2.columns = np.concatenate((np.array(['vendor_code']), table2.columns[1 : ]))
```

```python
table2
```

```python
table1
```

```python
# task 5: creating pivot tables

pivot_table1 = pd.DataFrame(table1.groupby(table1.columns[0]).count().iloc[:, 1]).reset_index()
pivot_table1.columns = ['vendor_code', 'count']

pivot_table2 = pd.DataFrame(table1.groupby(table1.columns[2]).count().iloc[:, 1]).reset_index()
pivot_table2.columns = ['name', 'count']

# entering data into excel file

wb = load_workbook('Excel_Zadania_Final_solutions.xlsx')
ws = wb.worksheets[-2]

ws.cell(row = 1, column = 1).value = 'vendor_code'
ws.cell(row = 1, column = 2).value = 'count'
ws.cell(row = 1, column = 4).value = 'name'
ws.cell(row = 1, column = 5).value = 'count'

for i, value in enumerate(pivot_table1.values):
    ws.cell(row = i + 2, column = 1).value = value[0]
    ws.cell(row = i + 2, column = 2).value = value[1]
    
for i, value in enumerate(pivot_table2.values):
    ws.cell(row = i + 2, column = 4).value = value[0]
    ws.cell(row = i + 2, column = 5).value = value[1]
    
wb.save("Excel_Zadania_Final_solutions.xlsx")
```

```python
# coloring cells for taks 1

duplicated_indexes = np.where(table1.iloc[ : , 2].duplicated(keep = False).values)[0]

wb = load_workbook('Excel_Zadania_Final_solutions.xlsx')
ws = wb.worksheets[-3]

fill_cell = PatternFill(patternType = 'solid', fgColor = 'C64747')

for index in duplicated_indexes:
    ws[f'D{index + 10}'].fill = fill_cell
    
wb.save('Excel_Zadania_Final_solutions.xlsx')
```

```python
# inserting values into excel file for tasks 2, 3 and 4

wb = load_workbook('Excel_Zadania_Final_solutions.xlsx')
ws = wb.worksheets[4]

for i, value in enumerate(table2.values):
    ws.cell(row = 10 + i, column = 6).value = value[0]
    ws.cell(row = 10 + i, column = 7).value = value[1]
    ws.cell(row = 10 + i, column = 8).value = value[2]
    ws.cell(row = 10 + i, column = 9).value = value[3]
    ws.cell(row = 10 + i, column = 10).value = value[4]
    ws.cell(row = 10 + i, column = 11).value = value[5]
    ws.cell(row = 10 + i, column = 12).value = value[6]
    ws.cell(row = 10 + i, column = 13).value = value[7]
    
for i, value in enumerate(names.values):
    ws[f'O{10 + i}'].value = value
    
wb.save("Excel_Zadania_Final_solutions.xlsx")
```

## exercise 4

```python
# load data from sheet

sheet_data = pd.read_excel('Excel_Zadania_Final.xlsx', 'Excercise 4_Zadanie 4')
sheet_data.columns = [i for i in range(len(sheet_data.columns))]
```

```python
sheet_data.head(10)
```

```python
table1 = sheet_data.iloc[5 : , 1 : 3].reset_index(drop = True)
table2 = pd.DataFrame(sheet_data.iloc[5 : , 4]).reset_index(drop = True)

table1.columns = ['code', 'turnover']
table2.columns = ['code']
```

```python
table1.head()
```

```python
table2.head()
```

```python
# creating column count_of_code

count_of_code = table1[table1.code != 'Average '].groupby('code').count() 
count_of_code.columns = ['count_of_code']
```

```python
count_of_code.head()
```

```python
# adding count_of_code column to table2

table2_sorted = table2.sort_values('code').reset_index()
table2 = table2_sorted.set_index('code').join(count_of_code).sort_values('index').drop('index', axis = 1).reset_index()
```

```python
table2.head()
```

```python
# creating total_turnover column

total_turnover = table1[table1.code != 'Average '].groupby('code').sum()
total_turnover.columns = ['total_turnover']
```

```python
total_turnover.head()
```

```python
# adding total_turnover column to table2

table2_sorted = table2.sort_values('code').reset_index()
table2 = table2_sorted.set_index('code').join(total_turnover).sort_values('index').drop('index', axis = 1).reset_index()

table2.columns = ['code', 'count_of_code', 'total_turnover']
```

```python
table2.head()
```

```python
# entering data to excel file

wb = load_workbook('Excel_Zadania_Final_solutions.xlsx')
ws = wb.worksheets[-1]

for i, values in enumerate(table2.values):
    ws.cell(row = i + 7, column = 6).value = values[1]
    ws.cell(row = i + 7, column = 7).value = values[2]
    
wb.save('Excel_Zadania_Final_solutions.xlsx')
```
